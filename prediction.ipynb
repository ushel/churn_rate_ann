```python
import pandas as pd
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
# load all the train model, scalar pickle,onehot
model = load_model('model.h5')

# load encoders and scaler

with open('onehot_encoder_gro.pkl','rb') as file:
    onehot_encoder_gro = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scalar.pkl','rb') as file:
    scaler = pickle.load(file) 
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded = onehot_encoder_gro.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_gro.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
#Encode categorical variable
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
# Concatination with onehot encoded data 
input_df = pd.concat([input_df.drop('Geography',axis=1),geo_encoded_df],axis=1)
input_df
```

```python
#Scaling data

input_scaled = scaler.transform(input_df)
input_scaled
```

```python
# Predict Churn
prediction = model.predict(input_scaled)
```

```python
prediction
```

```python
prediction_proba = prediction[0][0]
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print("The customer is not likely to churn.")
```
